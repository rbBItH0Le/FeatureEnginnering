```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('mercedesbenz.csv',usecols=['X1','X2','X3','X4','X5'])
```

```python
df
```

```python
for col in df.columns:
    print(col,':',len(df[col].unique()),'labels')
```

```python
df.X2.value_counts().sort_values(ascending=False)
```

```python
top_10=[x for x in df.X2.value_counts().sort_values(ascending=False).head(10).index]
```

```python
top_10
```

```python
for label in top_10:
    df[label]=np.where(df['X2']==label,1,0)
```

```python
df
```

```python
for col in df.columns:
    for label in top_10:
        df[col+'_'+label]=np.where(df[col]==label,1,0)
```
