```python
hu=pd.read_csv('titanic.csv',usecols=['Cabin','Age','Survived'])
```

```python
hu
```

```python
hu['Cabin'].fillna('Missing',inplace=True)
```

```python
hu['Cabin']=hu['Cabin'].astype('str')
```

```python
liso=[x[0] for x in hu['Cabin'].values]
```

```python
hu['encod']=liso
```

```python
hu
```

```python
lispo=list(hu.groupby('encod')['Survived'].mean().sort_values().index)
```

```python
dfier={k:i for i,k in enumerate(lispo)}
```
